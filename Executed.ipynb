```python
def transform_date(date):
    date_info = datetime.date.fromtimestamp(date)
    return date_info

#transform_date method changes the date format of film_date and published_date to YYYY-MM-DD
df_1['film_date'] = df_1['film_date'].apply(transform_date)
df_1['published_date'] = df_1['published_date'].apply(transform_date)


#this converts the duration column into minutes
df_1['duration'] = df_1['duration']/60

df_1.head()
```

```python
df_1['freq'] = df_1.groupby('main_speaker')['main_speaker'].transform('count')
df_1
```
